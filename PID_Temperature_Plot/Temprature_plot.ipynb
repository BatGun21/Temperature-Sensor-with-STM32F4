```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace the path below with the actual path to your log file
log_file_path = 'C:\\Users\\91742\\Desktop\\Vavye Mobility\\Projects\\PID_Temperature_Plot\\Temperature.log'

# Read the log file into a DataFrame
with open(log_file_path, 'r') as file:
    temperature_data = file.read().strip().split(',')

# Filter out empty strings and convert temperature data to a list of floats
temperature_values = [float(temp) for temp in temperature_data if temp]

# Create a DataFrame with time as the index and temperature values
df = pd.DataFrame(temperature_values, columns=['Temperature'])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Temperature'], marker='o', linestyle='-', color='b')
plt.title('Temperature Variation Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.show()

# Save the DataFrame to a CSV file (optional)
df.to_csv('temperature_data.csv', index=False)
```
